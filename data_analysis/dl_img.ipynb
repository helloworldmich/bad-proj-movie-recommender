```python
import requests
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'D:\tecky-academy\c17-bad-project-01-tw\data_src\meta_Movies_and_TV.csv')
df.head()
```

```python
df_img = df[["asin", "imageURL","imageURLHighRes"]]
df_img.head()
```

```python
for index, row in df.iterrows():
    url = df_img.iloc[index]['imageURLHighRes'][2:-2]
    filename = df_img.iloc[index]['asin'] + ".jpg"
    if url == "":
        continue 
    img_data = requests.get(url).content
    with open(filename, 'wb') as handler:
        handler.write(img_data)
```
